<h1>Pima Anomaly Detection using Isolation Forests Diabetes Usecase</h1>

<font color="darkblue">Isolation Forest is an unsupervised techniques of detecting anomalies. Here, We have an usecase on Diabetes with features such as Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age, Feature1 and Feature2</font>

### <font color=darkblue>To jump start, Use UMOYA (Repo).</font> 
UMOYA does versioning resources (like model, data and code) and manages its dependencies for ai projects.
MLW's Solution Team publishes Data, Model and Code for this Usecase on UMOYA (Repo) so User(s) can easily grab them all in one click.

<br/></br> There are two ways to download resource(s).
1. Download resource(s) from an MLW tenant UI.
<ol>
    <li>Go to "Repo" Tab.</li>
    <li>Select "Diabetes.pmml" </li>
    <li>Click on "Download" </li>
</ol>
2. Use UMOYA CLI.
Go to ZMOD resource directory in terminal, Give command : <br/>umoya add Diabetes.pmml

### <font color=darkred>&#187;</font> Test Diabetes Model :
1. Load "AnamolyDetectionUsingIsoFor.pmml" to ZMK.
2. Select "AnamolyDetectionUsingIsoFor-Test.csv" in "Data" Tab.
3. Click on "Predect Data".
4. Selct Model "AnamolyDetectionUsingIsoFor.pmml" and submit.

It will create new file with column "target" and value with 1 or -1 to indecate that respective patient has diabetes or not.
