<h1>DistractedDriver Usecase</h1>

<font color="darkblue">How can I predict and send alarm to Cumulocity When Driver is not driving safely; He/She is distracted with various activities.</font>


<font color=darkgreen>You can use Cumulocity Machine Learning Workbench (MLW) to resolve this problem</font>
<br/> How ? We can follow the below steps :
1. Create model from mobilenet and standard template(s).
2. Associate the python script as the post processing script to send alarm.
3. Train and validate with data.
4. Load DistractedDriver Model to MLW / ZMK execution engine.
5. Load Test image (png) in MLW.
6. Predict with the Model.
7. View json output file / Cumulocity alarm.

### <font color=darkblue>To jump start, Use UMOYA (Repo).</font> 
UMOYA does versioning resources (like model, data and code) and manages its dependencies for ai projects.
MLW Team published Data, Model and Code for Distracted Driver Usecase on UMOYA (Repo) so User can easily get them all in one click.

<br/></br> There are two ways to download resource(s).
1. Download resource(s) from an MLW tenant UI.
2. Use UMOYA CLI.

#### 1. Download resources from an MLW tenant UI.
<ol>
    <li>Go to "Repo" Tab.</li>
    <li>Select "DistractedDriver.pmml" </li>
    <li>Click on "Download" </li>
</ol>


#### 2. Use UMOYA CLI to add resource(s).
<ol>
    <li>Make sure you have UMOYA CLI installed on your interested directory. If not then Install <a href="https://github.com/Umoya-ai/UMOYA/tree/master/docs/cli">here</a></li>
    <li>Initialize your directory : umoya init</li>
    <li>Configure to hub.umoya.ai : umoya info -u https://hub.umoya.ai/v3/index.json </li>
    <li>Add Model and its dependent resource(s) : umoya add DistractedDriver.pmml</li>

Above Step 1 or 2 can help to get Model, Data and Code for Usecase.
To post alarm when scoring, PMML model needs to associate DistractedDriver.py script as post processing script and configure username and password in script.

For further query or help, You can visit https://mlw.ai
