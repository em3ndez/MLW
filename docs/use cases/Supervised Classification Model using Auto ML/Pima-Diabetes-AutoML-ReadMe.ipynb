<h1>Pima Diabetes Usecase</h1>

<font color="darkblue">How can We predict Diabetes When We know patient's information such as Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction and Age ?</font>


<font color=darkgreen>This is a supervised classification problem. We can use AutoML in Cumulocity IoT Machine Learning Workbench (MLW) to resolve this problem</font>
### <font color=darkred>&#187;</font> Train and Build Model with Auto ML :
1. Get train data-set in "Data" Tab.
2. Click on "Classification" and swtich on "Supervised"
3. On "Target" Column, Unselect "Use for Model" and select "Target" variable.
4. Click on "Build"
5. Give Model Name.
6. Select "Accuracy" in "Scoring"
7. Select "All" in "Algorithm"
8. Submit Training Parameter.

This will start process to find the best model with accuracy score. We can save model into "Models" Tab.

### <font color=darkblue>To jump start, Use UMOYA (Repo).</font> 
UMOYA does versioning resources (like model, data and code) and manages its dependencies for ai projects.
MLW Team published Data, Model and Code for this Usecase on UMOYA (Repo) so User can easily grab them all in one click.

<br/></br> There are two ways to download resource(s).
1. Download resource(s) from an MLW tenant UI.
<ol>
    <li>Go to "Repo" Tab.</li>
    <li>Select "Diabetes.pmml" </li>
    <li>Click on "Download" </li>
</ol>
2. Use UMOYA CLI.
Go to ZMOD resource directory in terminal, Give command : <br/>umoya add Diabetes.pmml

### <font color=darkred>&#187;</font> Test Diabetes Model :
1. Load "Diabetes.pmml" to ZMK.
2. Select "Diabetes_test.csv" in "Data" Tab.
3. Click on "Predect Data".
4. Selct Model "Diabetes.pmml" and submit.

It will create new file with column "target" and value with 1 or 0 to indecate that respective patient has diabetes or not.
